```python
import pandas as pd
import json
import os
import numpy as np
from scipy.stats import chi2_contingency
import requests
```

```python
file_list_fii = '../dataset/BDovI.csv'
path_cvm_mensal_records = '../dataset/cvm_mensal_records.json'
path_cvm_trimestral_records = '../dataset/cvm_trimestral_records.json'
path_cvm_dividend_records = '../dataset/cvm_dividend_records.json'
folder_cvm_mensal = 'cvm_mensal'
folder_cvm_trimestral = 'cvm_trimestral'
folder_cvm_dividend = 'cvm_dividend'
```

```python
df_list_fii = pd.read_csv(file_list_fii)
df_list_fii = df_list_fii.iloc[:, 2:]
df_list_fii
```

```python
pasta = f'../dataset/{folder_cvm_dividend}'
caminhos = [os.path.join(pasta, nome) for nome in os.listdir(pasta)]
arquivos = [os.path.basename(arq).split('_')[0] for arq in caminhos if os.path.isfile(arq)]
#print(arquivos)
arquivos
```

```python
df_documents = pd.DataFrame({"ticker": arquivos})
df_document_count = df_documents["ticker"].value_counts().reset_index()
df_document_count.columns = ["ticker", "count"]

df_document_count
```

```python
with open(path_cvm_dividend_records, 'r') as file:
    data = json.load(file)

df_records = pd.DataFrame(data)
df_records
```

```python
df_join = df_records.join(df_document_count.set_index('ticker'), on='ticker')
df_join[(df_join['recordsTotal'] != df_join['count']) & (~df_join['count'].isnull())]
```
